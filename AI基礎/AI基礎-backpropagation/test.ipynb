```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10, 10, 100)
```

```python
feature = torch.from_numpy(X.astype(np.float32))
a = torch.nn.ReLU()
out = a(feature)
plt.plot(X, out)
```

```python
feature = torch.from_numpy(X.astype(np.float32))
a = torch.nn.Sigmoid()
out = a(feature)
plt.plot(X, out)
```

```python
feature = torch.from_numpy(X.astype(np.float32))
a = torch.nn.Tanh()
out = a(feature)
plt.plot(X, out)
```
