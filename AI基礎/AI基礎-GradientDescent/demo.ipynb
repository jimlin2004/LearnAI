```python
x = 9.9
lr = 0.1
for i in range(20):
    grad = 2 * x
    x -= lr * grad
    print(x)
```

```python
def f(x):
    return (x**2)

lineX = np.linspace(-10, 10, 100)
dotX = 9.9
lr = 0.1
dotXData = [[dotX]]
dotYData = [[f(dotX)]]
for i in range(20):
    grad = 2 * dotX
    dotX -= lr * grad
    dotXData.append([dotX])
    dotYData.append([f(dotX)])

fig = plt.figure()
ax = fig.gca()

ax.plot(lineX, f(lineX))

dot, = ax.plot([], [], color = "red", marker = 'o', markersize = 10, linestyle = '')

def initDot():
    dot.set_data(dotXData[0], dotYData[0])
    return dot,
def updateDot(i):
    dot.set_data(dotXData[i], dotYData[i])
    return dot,

ani = animation.FuncAnimation(fig = fig, func = updateDot, init_func = initDot, frames = 20, interval = 1000/20, blit = True, repeat = True)
plt.show()
ani.save("gradientDescent1.gif", writer = "ffmpeg", fps=10)
```

```python
X = np.linspace(-10, 10, 100)
plt.plot(X, f(X))
```

```python
def f(x):
    return -30 * np.sin(x) + 2 * (x ** 2) - 5 * x
def f_dev(x):
    return -30 * np.cos(x) + 4 * x - 5

lineX = np.linspace(-10, 10, 100)
dotX = -9.9
lr = 0.08
N = 50
dotXData = [[dotX]]
dotYData = [[f(dotX)]]
for i in range(N):
    grad = f_dev(dotX)
    dotX -= lr * grad
    dotXData.append([dotX])
    dotYData.append([f(dotX)])

fig = plt.figure()
ax = fig.gca()

ax.plot(lineX, f(lineX))

dot, = ax.plot([], [], color = "red", marker = 'o', markersize = 10, linestyle = '')

def initDot():
    dot.set_data(dotXData[0], dotYData[0])
    return dot,
def updateDot(i):
    dot.set_data(dotXData[i], dotYData[i])
    return dot,

ani = animation.FuncAnimation(fig = fig, func = updateDot, init_func = initDot, frames = N, interval = 5000 / N, blit = True, repeat = True)
plt.show()
ani.save("gradientDescent2.gif", writer = "ffmpeg", fps=3)
```
